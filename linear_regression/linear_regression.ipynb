```python
from linear_regression import *
```

```python
csv_name = 'lake.csv'
df = make_plot(csv_name)
```

```python
X = []
for val in df.year:
    X.append(np.transpose(np.array([1, val])))
X = np.array(X)
Y = np.array(df.days)
print(X.shape)
print(Y.shape)
```

```python
hat_beta = train(X, Y)
hat_beta
```

```python
# predict the number of frozen days in the year 2023
predict(hat_beta, np.array([1, 2023]))
```

```python
# plot the regression
X_pre = np.arange(1855, 2030, 10)
Y_pre = hat_beta[0] + X_pre * hat_beta[1]
plt.plot(df.year, df.days)
plt.xticks(df.year)
plt.xlabel("Year")
plt.ylabel("Number of frozen days")
plt.locator_params(axis='y', nbins=10)
plt.locator_params(axis='x', nbins=10)
plt.plot(X_pre, Y_pre)
plt.savefig("regression.jpg")
```
